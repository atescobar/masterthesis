# Surface charge density at the interface



```python
import numpy as np

z = 2
e = 1.60217662E-19
k = 1.38064852E-23
T = 300
b=1/(k*T)
Na = 6.02E23
Fa = Na * e
q = e
R = Na * k
coef = z*e/(k*T)
V_0 = -coef*0.15
Cb1 = 0.1*1e-3
Cb2 = 0.5*1e-3
Cb3 = 0.2*1e-3

epsilon = 80.9 * 8.85418782E-12
k1 = np.sqrt(0.1 * (z*Fa) ** 2 * Cb1/ (R * T * epsilon))
k2 = np.sqrt(0.1 * (z*Fa) ** 2 * Cb2/ (R * T * epsilon))
k3 = np.sqrt(0.1 * (z*Fa) ** 2 * Cb3/ (R * T * epsilon))


```

```python
import matplotlib.pyplot as plt
import matplotlib



fig, ax = plt.subplots()
fs =24 
font = {'size'   : fs}
matplotlib.rc('font', **font)


scale = 1e2
V = np.arange(-0.4,0.4, 0.0001)
sig1 = -4*epsilon*k1/(b*q) * scale * np.tanh(q*b*V/4)/(np.tanh(q*b*V/4) ** 2 - 1)
sig2 = -4*epsilon*k2/(b*q) * scale * np.tanh(q*b*V/4)/(np.tanh(q*b*V/4) ** 2 - 1)
sig3 = -4*epsilon*k3/(b*q) * scale * np.tanh(q*b*V/4)/(np.tanh(q*b*V/4) ** 2 - 1)


plt.figure(1)
plt.style.use('thesis')
plt.title('', fontweight='bold')
#plt.title('Charge Density In Terms Of The Voltage At The Electrode', fontweight='bold')
plt.xlabel(r'Voltage [$V$]')
plt.ylabel(r'Charge Density [$V/m$] $\times 10^{-2}$')
plt.plot(V, sig1, 'b--', label='$C_b = '+str(1e3 * Cb1)+'M$')
plt.plot(V, sig2, 'g--', label='$C_b = '+str(1e3 * Cb2)+'M$')
plt.plot(V, sig3, 'r--', label='$C_b = '+str(1e3 * Cb3)+'M$')
plt.legend()
legend = ax.legend(loc='lower right')
# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('#FFFFFF')
plt.savefig('../../../img/sigma-voltage.eps',dpi=1000)
plt.show()
```

```python

sig = np.arange(-5,5, 0.0001)
tanhp1 = (-2*epsilon*k1 + np.sqrt((epsilon * k1) ** 2 + (b * q * sig) **2))/(b * q * sig)
tanhp2 = (-2*epsilon*k2 + np.sqrt((epsilon * k2) ** 2 + (b * q * sig) **2))/(b * q * sig)
tanhp3 = (-2*epsilon*k3 + np.sqrt((epsilon * k3) ** 2 + (b * q * sig) **2))/(b * q * sig)

tanhm1 = (-2*epsilon*k1 + np.sqrt((epsilon * k1) ** 2 - (b * q * sig) **2))/(b * q * sig)
tanhm2 = (-2*epsilon*k2 + np.sqrt((epsilon * k2) ** 2 - (b * q * sig) **2))/(b * q * sig)
tanhm3 = (-2*epsilon*k3 + np.sqrt((epsilon * k3) ** 2 - (b * q * sig) **2))/(b * q * sig)

Vp1 = 4/(q*b) * np.arctanh(tanhp1)
Vp2 = 4/(q*b) * np.arctanh(tanhp2)
Vp3 = 4/(q*b) * np.arctanh(tanhp3)

Vm1 = 4/(q*b) * np.arctanh(tanhm1)
Vm2 = 4/(q*b) * np.arctanh(tanhm2)
Vm3 = 4/(q*b) * np.arctanh(tanhm3)


fig, ax = plt.subplots(figsize=(20,16))
plt.grid(True, color= '#F2F2F2')
fs =24 
font = {'size'   : fs}
matplotlib.rc('font', **font)



plt.xlabel(r'Surface Electric Field ($V/m$)')
plt.ylabel(r'Voltage ($V$)')
plt.plot(sig, Vp1, 'g--', label = "$C_b = 0.1$")
plt.plot(sig, Vp2, 'b--', label = "$C_b = 0.5$")
plt.plot(sig, Vp3, 'r--', label = "$C_b = 0.01$")
plt.legend()
legend = ax.legend(loc='lower right', shadow=True, fontsize='x-large')
# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('#FFFFFF')
plt.savefig('../../../img/voltage-sigma.eps', format='eps', dpi=1000, fontsize=fs, fontweight='bold')
plt.show()
```
